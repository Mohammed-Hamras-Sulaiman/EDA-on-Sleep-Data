# Physical Activity VS Sleep Quality

```python
plt.figure(figsize=(12,6))
sns.barplot(x = 'Gender', y = 'Sleep Quality', data = df)
plt.title('sleep quality between males and females')
plt.xlabel('Gender')
plt.ylabel('Sleep Quality')
plt.show()
```

#### Females have better sleep as compare to Males

```python
#define custom codes
custom_codes = {'low' : 1, 'medium' : 2, 'high' : 3}
# apply the custom codes using map
df['Physical Activity Level']=df['Physical Activity Level'].map(custom_codes)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x = 'Gender', y = 'Physical Activity Level', data = df)
plt.title('Physical Activity Level between males and females')
plt.xlabel('Gender')
plt.ylabel('Physical Activity Level')
plt.show()
```

females do more activity than males so they recieve more sleep

```python
from scipy import stats

sleep_q_males = df[df['Gender'] == 'm']['Sleep Quality']
sleep_q_females = df[df['Gender'] == 'f']['Sleep Quality']

t_sq , p_sq = stats.ttest_ind(sleep_q_males, sleep_q_females, nan_policy='omit')

pal_males = df[df['Gender'] == 'm']['Physical Activity Level'] #pal : Physical Activity Level
pal_females = df[df['Gender'] == 'f']['Physical Activity Level']

t_pal , p_pal = stats.ttest_ind(pal_males, pal_females, nan_policy='omit')
```

```python
print(f"t-sleep-quality  + {t_sq:.2f}")
print(f"p-sleep-quality  + {p_sq:.2f}")
print(f"t-physical_activity  + {t_pal:.2f}")
print(f"p-physical_activity  + {p_pal:.2f}")
```

 The T-statistic of -14.32 and P-value of 0.0000 show that males and females have a statistically significant difference in their sleep quality, with males having a lower average sleep quality than females. For Physical Activity Level: "Is there a significant difference in the physical activity levels between males and females?" The T-statistic of -10.98 and P-value of 0.0000 again indicate a significant difference, with males having lower physical activity levels compared to females. Summary You found the T-values and P-values to compare the average sleep quality and physical activity levels between males and females. Since both P-values are close to 0, it means the differences you found are statistically significant, meaning males and females are likely to have different sleep quality and physical activity levels in the dataset.

#### Timing of Sleep and waking up vs Sleep Quality

```python
def sleep_duration_hours_helper_function(str_time):
    h = int(str_time.split(':')[0])
    m = int(str_time.split(':')[1])
    return h + m/60
    
df['Bedtime'] = df['Bedtime'].apply(sleep_duration_hours_helper_function)
df['Wake-up Time'] = df['Wake-up Time'].apply(sleep_duration_hours_helper_function)

df.head()
```

```python
df['Sleep Duration Hours'] = (df['Wake-up Time'] - df['Bedtime']) % 24
df.head()
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x = 'Bedtime', y = 'Sleep Quality', data = df)
plt.xlabel('Bed Time')
plt.ylabel('Sleep Quality')
plt.title('Bed Time VS Sleep Quality')
plt.show()
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x = 'Wake-up Time', y = 'Sleep Quality', data = df)
plt.xlabel('Wake-up Time')
plt.ylabel('Sleep Quality')
plt.title('Wake-up Time VS Sleep Quality')
plt.show()
```

```python
plt.figure(figsize = (12, 6))
sns.barplot(x = 'Sleep Duration Hours', y = 'Sleep Quality', data = df)
plt.xlabel("Sleep Duration Hours")
plt.ylabel("Sleep Quality")
plt.title("Sleep Duration Hours vs Sleep Quality")
plt.show()
```

checking Linear Relationship b\w Sleep Time Hours & Sleep Quality

```python
X = df[['Sleep Duration Hours']]
Y = df['Sleep Quality']
model = LinearRegression()
model.fit(X,Y)
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x='Sleep Duration Hours',y = 'Sleep Quality',data = df)
plt.show()
```

```python
plt.figure(figsize = (6,6))
sns.scatterplot(x = 'Sleep Duration Hours', y = 'Sleep Quality', data = df, label = 'Actual Data')
plt.plot(X,model.predict(X), color = 'red', label = 'Predicted Sleep Quality')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(Y, model.predict(X))
```
